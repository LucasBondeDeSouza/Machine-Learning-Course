```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
```

```python
with open('credit.pkl', 'rb') as f:
    x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

### Treinamento do Algorítmo

```python
naive_credit_data = GaussianNB()
naive_credit_data.fit(x_credit_treinamento, y_credit_treinamento)
```

### Previsão

```python
previsoes = naive_credit_data.predict(x_credit_teste)
previsoes
# 1 -> Não paga o empréstimo
# 0 -> Paga o empréstimo
```

### Comparar o base de teste com os registros reais

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_credit_teste, previsoes)
# Percentual de ACERTO
```

```python
confusion_matrix(y_credit_teste, previsoes)
# Informa a quantidade de acerto e erro dos clientes que pagam o emprestimo e dos clientes que não pagam o emprestimo
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(naive_credit_data)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```
